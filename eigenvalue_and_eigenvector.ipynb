```python
import numpy as np
import pandas as pd
```

```python
def det(a):
  d=a[0,0]*a[1,1]-a[0,1]*a[1,0]
  return d
```

```python
A=np.array([[3, 2], [1, 4]])
B=np.array([[2, 6], [1, 3]])
C=np.array([[0, 1], [-1, 0]])

print("The determinant of A:",det(A))
print("The determinant of A:",det(B))
print("The determinant of A:",det(C))
```

```python
A =np.array([[5, 2], [2, 5]])

vals, vecs = np.linalg.eig(A)

print(vals)  #eigenvalues

print(vecs)  # eigenvectors
#checking A @ v = lambda * v
v = vecs[:, 0]

print(np.allclose(A @ v, vals[0] * v))
```
