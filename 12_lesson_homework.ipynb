Qo'lda hisoblang: E_X = n*p, Var_X = n*p*(1-p)

```python
n,p =20,0.3
# print( f"Var: {(n * p * (1 - p)).round(2)}, E[x]: {n*p}, Standard deviation:{round(np.sqrt(n * p * (1 - p)), 2)}")
print(f"Var: {round(n * p * (1 - p), 2)}, E[x]: {n*p}, Standard deviation: {round(np.sqrt(n * p * (1 - p)), 2)}")
dist=stats.binom(n, p)
print(f"Var:{dist.var()}")
print(f"E[X]: {dist.mean()}")
print(f"standard deviation: {dist.std().round(2)}\n")

x = np.arange(0, 21)
plt.bar(x, dist.pmf(x), color="blue", alpha=0.9, ec='yellow')
plt.title("Binomial graph")
plt.xlabel('X')
plt.ylabel('P(X=x)')
plt.ylim(0, 0.2)
plt.axvline(dist.mean(),color='pink',linestyle='--')
plt.axvline(dist.mean()-dist.std(),color='red',linestyle='--')
plt.axvline(dist.mean()+dist.std(),color='red',linestyle='--')
plt.show()
```

4.Noto'g'ri Zar — PMF Tekshiruvi



```python
x=np.array([1,2,3,4,5])
probs = np.array([0.10, 0.20, 0.30, 0.25, 0.20])

# 1. Yig'indini tekshiring
print(f"Yig'indi: {sum(probs)}")  # 1 bo'lishi kerak

# 2. P(X=5) ni tuzating
probs[4] = round(1 - sum(probs[:4]),2)
print(probs)

# 3. E[X] ni hisoblang va grafik chizing
E_X = np.sum(x * probs)
print(f"E[X] = {E_X:.2f}")

plt.bar(x, probs,
        color='blue',
        edgecolor='black',
        ls=':',
        lw=0.5,
        width=0.6,
        )

plt.title("Probability mass function")
plt.xlabel("X")
plt.ylabel("P(X)")
plt.show()
```

5.Poisson Taqsimoti — Mijozlar Oqimi



```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

lam = 5
dist = stats.poisson(lam)

print(f"E[X] = {dist.mean()}, Var(X) = {dist.var()}")

print(f"P(X=0) = {dist.pmf(0):.4f}")

print(f"P(X>=8) = {dist.sf(7):.4f}")

x = np.arange(0, 16)
y = dist.pmf(x)

plt.figure(figsize=(8,5))
plt.bar(x, y, color="#2ECB9B", alpha=0.85)

plt.axvline(lam, color='red', linestyle='--', label=f'E[X]={lam}')

plt.title("Poisson Distribution (λ=5)")
plt.xlabel("Number of customers")
plt.ylabel("Probability")
plt.xticks(x)
plt.legend()
plt.show()
```
