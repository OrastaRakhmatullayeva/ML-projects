```python
import numpy as np
import matplotlib.pyplot as plt
A = np.array([[2, 0],
              [0, 1]])
v = np.array([1, 1])
w = A @ v                    # [2, 1]
plt.quiver(0, 0, v[0], v[1], angles='xy',
           scale_units='xy', scale=1, color='gray')
plt.quiver(0, 0, w[0], w[1], angles='xy',
           scale_units='xy', scale=1, color='green')

plt.xlim(-1, 3); plt.ylim(-1, 3)
plt.grid(True); plt.show()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[0, -1],
              [1, 0]])
v = np.array([1, 0])
w = A @ v                     # [2, 1]

plt.quiver(0, 0, v[0], v[1], angles='xy',
           scale_units='xy', scale=1, color='gray')

plt.quiver(0, 0, w[0], w[1], angles='xy',
           scale_units='xy', scale=1, color='green')

plt.xlim(-2, 3); plt.ylim(-2, 3)
plt.grid(True); plt.show()
```

```python
M1 = np.array([[0, -1], [1, 0]])
print(M1 @ [1, 0])   # [0 1]

M2 = np.array([[2, 0], [0, 2]])
print(M2 @ [3, -1])  # [ 6 -2]

M3 = np.array([[1, 0], [0, -1]])
print(M3 @ [2, 3])   # [ 2 -3]

M4 = np.array([[1, 1], [0, 1]])
print(M4 @ [2, 1])   # [3 1]
```

```python
theta = np.radians(45)            # gradus -> radian!
R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])
print(R @ [1, 0])         # [0.707 0.707]
```

```python
# --- Aylantirish ---
theta = np.radians(45)            # gradus -> radian!
R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])

v_old = np.array([1, 0])          # oldingi vektor
v_new = R @ v_old                 # aylantirilgan vektor

# --- Grafik ---
fig, ax = plt.subplots(figsize=(6, 6))

# Oldingi vektor (kulrang)
ax.quiver(0, 0, v_old[0], v_old[1], angles='xy',
          scale_units='xy', scale=1, color='gray',
          width=0.012, label='Oldin  [1, 0]')

# Keyingi vektor (yashil)
ax.quiver(0, 0, v_new[0], v_new[1], angles='xy',
          scale_units='xy', scale=1, color='#2ECB9B',
          width=0.012, label='Keyin  [0.707, 0.707]')


# --- Bezaklar ---
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)
ax.grid(True, linestyle='--', alpha=0.4)
ax.legend(loc='upper right')
ax.set_title('Vektorni 45° ga aylantirish')
plt.show()
```

```python
# Kvadrat: 4 burchak, ustun-vektorlar sifatida
kv = np.array([[0, 1, 1, 0, 0],
               [0, 0, 1, 1, 0]])

Tlar = {'Scaling':    [[2, 0], [0, 0.5]],
        'Rotation':   [[0, -1], [1, 0]],
        'Reflection': [[1, 0], [0, -1]],
        'Shear':      [[1, 1], [0, 1]]}

fig, axes = plt.subplots(1, 4, figsize=(14, 3.5))
for ax, (nom, T) in zip(axes, Tlar.items()):
    yangi = np.array(T) @ kv
    ax.plot(kv[0], kv[1], 'gray')
    ax.plot(yangi[0], yangi[1], 'green')
    ax.set_title(nom); ax.grid(False); ax.axis('equal')
```

```python
S = np.array([[2, 0], [0, 1]])   # cho'zish
R = np.array([[0, -1], [1, 0]])  # 90° burish
v = np.array([1, 0])

print(R @ S)        # [[0 -1], [2 0]]
print(S @ R)        # [[0 -2], [1 0]]

print(R @ S @ v)    # [0 2]  avval cho'z, keyin bur
print(S @ R @ v)    # [0 1]  avval bur, keyin cho'z

print(np.allclose(R @ S, S @ R))  # False!
```

```python
def f(img):
  versiyalar = {'Asl':        img,
                'fliplr':     np.fliplr(img),
                'flipud':     np.flipud(img),
                'rot90':      np.rot90(img)}

  fig, axes = plt.subplots(2, 2, figsize=(9, 6))
  for ax, (nom, v) in zip(axes.flat, versiyalar.items()):
      ax.imshow(v)
      ax.set_title(nom)
      ax.axis('off')
  plt.show()
```
