```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
titanic=sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
titanic.isnull().sum().sort_values(ascending=False).head()
```

```python
num = titanic.select_dtypes(include='number').columns
titanic[num] = titanic[num].fillna(titanic[num].median())

cat = titanic.select_dtypes(include='object').columns
titanic[cat] = titanic[cat].fillna('None')
```

```python
titanic.isnull().sum().head()
```

SCATTERPLOT

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=titanic,
                x='age',
                y='fare',
                hue='class')

plt.title("Relationship betweeen age and fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title='Class')
plt.show()
```

PEARSON R

```python
from scipy import stats
r, p_val = stats.pearsonr(titanic['age'], titanic['fare'])
print(f"r={r:.4f}, p={p_val:.6f}")
```

```python
r_normal = np.corrcoef(diamond['carat'], diamond['price'])[0, 1]
print(f"Oddiy Pearson r: {r_normal:.4f}")

r_log = np.corrcoef(np.log1p(diamond['carat']), np.log1p(diamond['price']))[0, 1]
print(f"Log transform qilingan Pearson r: {r_log:.4f}")
```

Topshiriq 3 — Diamonds: Korrelyatsiya Heatmap

```python
diamond = sns.load_dataset('diamonds')
numeric_cols = ['carat', 'depth', 'table', 'price', 'x', 'y', 'z']
```

CORRELATION MATRIX

```python
corr_matrix=diamond[numeric_cols].corr()
```

HEATMAP

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix,
            annot=True,
            fmt='.2f',
            cmap='RdYlGn',
            vmin=-1,
            vmax=1
)
plt.title("Diamonds Dataset — Correlation Matrix")
plt.show()
```

```python
corr_with_price = diamond[numeric_cols].corr()['price'].sort_values(ascending=False)
print(corr_with_price)
```

TODO 5c:  Pearson va Spearman r — taqqoslang


```python
from scipy import stats

r_pearson, p_pearson = stats.pearsonr(flight['year'], flight['passengers'])
print(f"Pearson r = {r_pearson:.4f}, p = {p_pearson:.6f}")

r_spearman, p_spearman = stats.spearmanr(flight['year'], flight['passengers'])
print(f"Spearman r = {r_spearman:.4f}, p = {p_spearman:.6f}")

print(f"\nFarq: {abs(r_pearson - r_spearman):.4f}")
if abs(r_pearson - r_spearman) < 0.05:
    print("Almost in same position.")
else:
    print("there is a difference")
```

# TODO 5d: Eng ko'p yo'lovchi bo'lgan oy va eng kam bo'lgan oyni toping


```python
max_row = flight.loc[flight['passengers'].idxmax()]
min_row = flight.loc[flight['passengers'].idxmin()]

print(f"Maximal passengers: {max_row['month']} {max_row['year']} — {max_row['passengers']} ta")
print(f"Minimal passengers: {min_row['month']} {min_row['year']} — {min_row['passengers']} ta")

plt.figure(figsize=(10, 6))
sns.scatterplot(data=flight, x='year', y='passengers', color='gray', alpha=0.5, label='Boshqa oylar')

plt.scatter(max_row['year'], max_row['passengers'], color='green', s=150, label='Max', zorder=5)
plt.scatter(min_row['year'], min_row['passengers'], color='red', s=150, label='Min', zorder=5)

plt.title("Passengers during years")
plt.xlabel("Year")
plt.ylabel("Number of passengers")
plt.legend()
plt.show()
```
